```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
data = pd.read_csv('University.csv')

# Group the data by 'Part' and calculate the count
part_counts = data['Part'].value_counts()

# Convert Series to DataFrame
part_counts = pd.DataFrame({'Part': part_counts.index, 'Count': part_counts.values})

# Create a pie chart
plt.pie(part_counts['Count'], labels=part_counts['Part'], autopct='%1.1f%%')
plt.title("Distribution of Parts")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Display the chart
plt.show()
```

```python
# Group the data by 'Part' and calculate the sum of 'Frequency'
part_frequency = data.groupby('Part')['Frequency'].sum()

# Create a pie chart
plt.pie(part_frequency, labels=part_frequency.index, autopct='%1.1f%%')
plt.title("Distribution of Frequency by Part")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Display the chart
plt.show()
```

```python
# Group the data by 'Region' and calculate the sum of 'Frequency'
region_frequency = data.groupby('Region')['Frequency'].sum()

# Sort the regions based on frequency in descending order
sorted_regions = region_frequency.sort_values(ascending=False)

# Select the top 5 regions
top_regions = sorted_regions[:5]

# Calculate the frequency sum of the remaining regions
other_frequency = sorted_regions[5:].sum()

# Create a list of labels for the top 5 regions
top_labels = top_regions.index.tolist()

# Add the 'Other' label for the remaining regions
top_labels.append('Other')

# Calculate the frequencies for the top 5 regions and the remaining regions
top_frequencies = top_regions.tolist()
top_frequencies.append(other_frequency)

# Create a pie chart
plt.pie(top_frequencies, labels=top_labels, autopct='%1.1f%%')
plt.title("Distribution of Frequency by Region")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Display the chart
plt.show()
```

```python
# Group the data by 'Region' and calculate the sum of 'Frequency'
region_frequency = data.groupby('Region')['Frequency'].sum()

# Create a pie chart
plt.pie(region_frequency, labels=region_frequency.index, autopct='%1.1f%%')

# Add count data as text inside each slice
count_data = region_frequency.values
percentage_data = region_frequency.values / sum(count_data) * 100

pie_labels = [f"{label} ({count}, {percentage:.1f}%)"
              for label, count, percentage in zip(region_frequency.index, count_data, percentage_data)]
plt.gca().set_aspect("equal")
plt.gca().set_title("Distribution of Frequency by Region")

plt.legend(pie_labels, loc='best')

# Display the chart
plt.show()
```

```python
# Group the data by 'Region' and calculate the sum of 'Frequency'
region_frequency = data.groupby('Region')['Frequency'].sum()

# Calculate percentage
percentage_data = region_frequency / region_frequency.sum() * 100

# Create a table DataFrame
table_data = pd.DataFrame({
    'Region': region_frequency.index,
    'Frequency': region_frequency.values,
    'Percentage': percentage_data.values.round(1)
})

# Sort the table in descending order by 'Frequency'
table_data = table_data.sort_values(by='Frequency', ascending=False)

# Reset the index of the table
table_data = table_data.reset_index(drop=True)

# Save the table data to a CSV file
table_data.to_csv('regiondatanopart.csv', index=False)

# Print the sorted table
print(table_data)
```

```python
# Group the data by 'Part' and 'Region' and calculate the sum of 'Frequency'
part_region_frequency = data.groupby(['Part', 'Region'])['Frequency'].sum()

# Calculate percentage within each part
percentage_data = part_region_frequency.groupby('Part').apply(lambda x: x / x.sum() * 100)

# Create a table DataFrame
table_data = pd.DataFrame({
    'Part': part_region_frequency.index.get_level_values('Part'),
    'Region': part_region_frequency.index.get_level_values('Region'),
    'Frequency': part_region_frequency.values,
    'Percentage': percentage_data.values.round(1)
})

# Sort the table in descending order by 'Frequency' within each part
table_data = table_data.sort_values(by=['Part', 'Frequency'], ascending=[True, False])

# Reset the index of the table
table_data = table_data.reset_index(drop=True)

# Print the sorted table
print(table_data)

# Save the table data to a CSV file
table_data.to_csv('regiondata.csv', index=False)

# Print a message to confirm successful saving
print("Table data saved to 'sorted_table_data.csv' file.")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you have your data in a pandas DataFrame called 'df'
school_types = data['Type'].values.tolist()

private_count = school_types.count('Private')
public_count = school_types.count('Public')
total_count = private_count + public_count

private_percentage = (private_count / total_count) * 100
public_percentage = (public_count / total_count) * 100

labels = ['Private', 'Public']
sizes = [private_percentage, public_percentage]
colors = ['blue', 'green']
explode = (0.1, 0)  # To emphasize the 'Private' slice, you can explode it

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle
plt.title('Distribution of Schools by Type')
plt.show()
```
