```python
import csv
import requests
from bs4 import BeautifulSoup

# Send a GET request to the URL
url = "https://www.thesummitexpress.com/2019/05/top-10-passers-secondary-march-2019-let-result.html"
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table on the page
table = soup.find('table')

# Extract data from the table
data = []
for row in table.find_all('tr'):
    # Skip header row
    if row.find('th'):
        continue
    
    # Extract data from each column in the row
    columns = row.find_all('td')
    row_data = [column.text.strip() for column in columns]
    
    # Append row data to the result
    data.append(row_data)

# Save data to a CSV file
filename = "rob.csv"
with open(filename, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(data)

print(f"Data saved to {filename} successfully.")


for row in data:
    print(row)
```

```python
# Step 1: Read the data
data = pd.read_csv('Book3.csv')

# Step 2: Filter the top 10 schools
top_schools = data[data['Rank'] <= 10]['School']

# Step 3: Count the frequency of schools
school_counts = top_schools.value_counts()

# Step 4: Create a DataFrame with school frequencies
top_performing_schools = pd.DataFrame({'School': school_counts.index, 'Frequency': school_counts.values})

# Step 5: Print the results in table form
print(top_performing_schools)

top_performing_schools.to_csv('top_performing_schools.csv', index=False)
```

```python
import matplotlib.pyplot as plt

grouped_data = data.groupby(['Month', 'Year'])['Rating'].mean().reset_index()

examination_dates = ['March 2018', 'September 2018', 'March 2019', 'September 2019', 'September 2021', 'January 2022', 'March 2022', 'June 2022', 'October 2022', 'March 2023']
grouped_data['Examination'] = pd.Categorical(grouped_data['Month'] + ' ' + grouped_data['Year'].astype(str), categories=examination_dates, ordered=True)
grouped_data = grouped_data.sort_values('Examination')

plt.plot(grouped_data['Examination'], grouped_data['Rating'], marker='o')
plt.xlabel('Examination')
plt.ylabel('Average Rating')
plt.title('Average Rating per Examination')
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python
grouped_data = data.groupby(['Month', 'Year', 'Level'])['Rating'].mean().reset_index()

examination_dates = ['March 2018', 'September 2018', 'March 2019', 'September 2019', 'September 2021', 'January 2022', 'March 2022', 'June 2022', 'October 2022', 'March 2023']
grouped_data['Examination'] = pd.Categorical(grouped_data['Month'] + ' ' + grouped_data['Year'].astype(str), categories=examination_dates, ordered=True)
grouped_data = grouped_data.sort_values(['Examination', 'Level'])

fig, ax = plt.subplots()

elementary_data = grouped_data[grouped_data['Level'] == 'Elementary']
secondary_data = grouped_data[grouped_data['Level'] == 'Secondary']

ax.plot(elementary_data['Examination'], elementary_data['Rating'], marker='o', label='Elementary')
ax.plot(secondary_data['Examination'], secondary_data['Rating'], marker='o', label='Secondary')

plt.xlabel('Examination')
plt.ylabel('Average Rating')
plt.title('Average Rating per Examination (Separate by Level)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

plt.show()
```
