```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('cleaneddata.csv')

# Assuming your dataset is stored in a pandas DataFrame called 'df'
df['dob'] = pd.to_datetime(df['dob'], format='%d/%m/%Y')

# Extract the birth year from the 'dob' column
birth_years = df['dob'].dt.year

# Define the birth year ranges for each generation
generation_ranges = {
    'Silent Generation': (1925, 1945),
    'Baby Boomers': (1946, 1964),
    'Generation X': (1965, 1980),
    'Millennials': (1981, 1996),
    'Generation Z': (1997, 2012),
}

# Categorize customers into generations based on their birth year
generation_counts = {generation: 0 for generation in generation_ranges}

for birth_year in birth_years:
    for generation, year_range in generation_ranges.items():
        if year_range[0] <= birth_year <= year_range[1]:
            generation_counts[generation] += 1
            break

# Convert the counts into a pandas Series for easier plotting
generation_series = pd.Series(generation_counts)

# Plot the pie chart
generation_series.plot(kind='pie', autopct='%.1f%%')
plt.axis('equal')
plt.ylabel('')
plt.title('Distribution of Customers by Generation')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

expense_categories = ['Entertainment', 'Food', 'HealthFitness', 'Home', 'KidsPets', 'PersonalCare', 'Grocery', 'Miscellaneous', 'Shopping', 'Travel']

# Define the generations and their corresponding birth year ranges
generations = {
    'Silent Generation': (1925, 1945),
    'Baby Boomers': (1946, 1964),
    'Generation X': (1965, 1980),
    'Millennials': (1981, 1996),
    'Generation Z': (1997, 2012),
}

# Create a list to store the expense lines for each category
expense_lines = []

for generation, (start_year, end_year) in generations.items():
    generation_df = df[(df['dob'].dt.year >= start_year) & (df['dob'].dt.year <= end_year)]
    
    expense_sums = generation_df[expense_categories].sum()
    
    # Store the expense sums as a line for the current generation
    expense_lines.append(expense_sums)

# Create the line graph
for category in expense_categories:
    expense_values = []
    
    # Iterate over the expense lines for each generation
    for line in expense_lines:
        expense_values.append(line[category])
    
    # Plot the line for the current expense category
    plt.plot(list(generations.keys()), expense_values, label=category)

# Set the axis labels and title
plt.xlabel('Generation')
plt.ylabel('Total Expenses')
plt.title('Expenses per Generation and Category')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Add a legend to the graph
plt.legend()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

expense_categories = ['Entertainment', 'Food', 'HealthFitness', 'Home', 'KidsPets', 'PersonalCare', 'Grocery', 
                      'Miscellaneous', 'Shopping', 'Travel']

# Define the generations and their corresponding birth year ranges
generations = {
    'Silent Generation': (1925, 1945),
    'Baby Boomers': (1946, 1964),
    'Generation X': (1965, 1980),
    'Millennials': (1981, 1996),
    'Generation Z': (1997, 2012),
}

# Create a new DataFrame to store the transposed data
transposed_df = pd.DataFrame(index=expense_categories)

# Iterate over the generations and their birth year ranges
for generation, (start_year, end_year) in generations.items():
    # Filter the DataFrame for the current generation
    generation_df = df[(df['dob'].dt.year >= start_year) & (df['dob'].dt.year <= end_year)]
    
    # Calculate the sum of expenses for each category in the current generation
    expense_sums = generation_df[expense_categories].sum()
    
    # Add the expenses for the current generation to the transposed DataFrame
    transposed_df[generation] = expense_sums

# Create the line graph
transposed_df.plot.line(marker='o')

# Set the axis labels and title
plt.xlabel('Expense Category')
plt.ylabel('Total Expenses')
plt.title('Expenses per Category for Each Generation')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
