```python
import pandas as pd

# Step 1: Read the CSV file
data = pd.read_csv('top_50_jobs_average_spending.csv')

# Step 2: Define the salary range criteria and convert to numerical values
salary_ranges = {
    "15,000 - 25,000": (15000, 25000),
    "25,000 - 40,000": (25000, 40000),
    "40,000 - 70,000": (40000, 70000),
    "70,000 - 120,000": (70000, 120000)
}

# Step 3: Function to categorize salary ranges
def categorize_salary_range(salary_range_str):
    lower, upper = salary_ranges[salary_range_str]
    return f"${lower}-{upper}"

# Step 4: Add the "Salary Range" column to the data
data['Salary Range'] = data['Average_Salary_Range'].apply(categorize_salary_range)

# Step 5: Group data by job and salary range, then calculate average spending
average_spending_by_job_and_salary = data.groupby(['job', 'Salary Range'])['Total_Spending'].mean().reset_index()

# Step 6: Pivot the data for better presentation
pivot_table = average_spending_by_job_and_salary.pivot(index='job', columns='Salary Range', values='Total_Spending')

# Step 7: Display the pivot table
print(pivot_table)
```

```python
import pandas as pd
import re

# Load the CSV data into a pandas DataFrame
df = pd.read_csv('top_50_jobs_average_spending.csv')

# Function to extract lower and upper bounds from the 'Average_Salary_Range' column
def extract_salary_range(salary_range):
    match = re.findall(r'\d+', salary_range)
    lower, upper = map(int, match)
    return f"{lower:,} - {upper:,}"

# Apply the function to create a new 'Salary_Range' column
df['Salary_Range'] = df['Average_Salary_Range'].map(extract_salary_range)

# Calculate the average total spending for each job salary range criteria
average_spending = df.groupby('Salary_Range')['Total_Spending'].mean()

# Print the results in the desired format
print("Average Salary Range | Average Total Spending")
print("----------------------------------------------")
for salary_range, avg_spending in average_spending.items():
    print(f"{salary_range:21} | {avg_spending:.2f}")
```

```python
import pandas as pd

# Load the CSV file
file_path = "top_50_jobs_average_spending.csv"
df = pd.read_csv(file_path)

# Convert 'Average_Salary_Range' column to numeric
df['Average_Salary_Range'] = df['Average_Salary_Range'].str.replace(',', '').str.extract(r'(\d+)', expand=False).astype(float)

# Define the job salary ranges
salary_ranges = [
    (15000, 25000),
    (25000, 40000),
    (40000, 70000),
    (70000, 120000)
]

# Calculate the average total spending for each job salary range
result = []
for salary_range in salary_ranges:
    lower_bound, upper_bound = salary_range
    mask = (df['Average_Salary_Range'] >= lower_bound) & (df['Average_Salary_Range'] < upper_bound)
    average_spending = df.loc[mask, 'Total_Spending'].mean()
    result.append((f"{lower_bound}-{upper_bound}", average_spending))

# Display the results
for range_str, average_spending in result:
    print(f"Salary Range: {range_str}")
    print(f"Average Total Spending: ${average_spending:.2f}")
    print()
```
