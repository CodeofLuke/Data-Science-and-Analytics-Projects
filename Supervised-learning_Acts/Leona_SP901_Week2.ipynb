```python
# Step 1: Load the data and import modules
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import RocCurveDisplay, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
cancer = load_breast_cancer()
```

```python
# Step 2: Split the data
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, test_size=0.3, random_state=42
)
```

```python
# Step 3: Initialize classifiers and train them
log_reg = LogisticRegression(max_iter=10000).fit(X_train, y_train)
rf = RandomForestClassifier(random_state=42).fit(X_train, y_train)
svm = SVC(probability=True).fit(X_train, y_train)  # Probability must be True to plot ROC curve
```

```python
# Step 4: Plot results for each classifier
classifiers = {'Logistic Regression': log_reg, 'Random Forest': rf, 'SVM': svm}
for name, clf in classifiers.items():
    RocCurveDisplay.from_estimator(clf, X_test, y_test)
    plt.title(f'ROC Curve for {name}')
    plt.show()

    ConfusionMatrixDisplay.from_estimator(clf, X_test, y_test)
    plt.title(f'Confusion Matrix for {name}')
    plt.show()
```

```python
# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, test_size=0.3, random_state=42
)
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Initialize the Logistic Regression classifier
log_reg = LogisticRegression(max_iter=10000)

# Train the classifier
log_reg.fit(X_train, y_train)

# Predict on the test set
log_reg_pred = log_reg.predict(X_test)

# Print the classification report
print("Logistic Regression:")
print(classification_report(y_test, log_reg_pred))
```

```python
# Initialize the Random Forest classifier
rf_clf = RandomForestClassifier(random_state=42)

# Train the classifier
rf_clf.fit(X_train, y_train)

# Predict on the test set
rf_pred = rf_clf.predict(X_test)

# Print the classification report
print("Random Forest:")
print(classification_report(y_test, rf_pred))
```

```python
# Initialize the Support Vector Machine classifier
svm_clf = SVC()

# Train the classifier
svm_clf.fit(X_train, y_train)

# Predict on the test set
svm_pred = svm_clf.predict(X_test)

# Print the classification report
print("Support Vector Machine (SVM):")
print(classification_report(y_test, svm_pred))
```
